```python
class ROI_Calc():

    def income(self):
        base_income = input("What is the rent for this?")
        add_income = input("Any additional income this property generates?")
        total_income = float(base_income) + float(add_income)
        self.monthly_income = "{:.2f}".format(total_income)
        print(f"\nTotal Monthly Income: ${self.monthly_income}.\n")


    def expenses(self):
        tax = input("What is the monthly tax owed for the property? ")
        insurance = input("What is the monthly cost of insurance for the property? ")
        utilities = ""
        utilities_q = input("Is the tenant responsible for utility bills? (Y/N) ")
        if utilities_q.lower() == "y":
            utilities = 0
        else:
            electric = input("What is average monthly electric bill? ")
            water = input("What is the average monthly water bill? ")
            sewer = input("What is the average monthly sewer bill? ")
            garbage = input("What is the average monthly garbage bill? ")
            gas = input("What is the average monthly gas bill? ")
            utilities = float(electric) + float(water) + float(sewer) + float(garbage) + float(gas)
        hoa = input("What is the monthly HOA fees? ")
        maintenance = input("What is the monthly cost of maintenance? ")
        vacancy = input("How much money is set aside each month to prepare for future vacancy? ")
        repairs = input("How much money is set aside each month for small repairs? ")
        capex = input("How much money is srt aside each month for capital expenditures? (big repairs/renovations, etc.) ")
        prop_man = input("How much money is spent on a property management team each month? ")
        mortgage = input("What is the monthly mortgage payment? ")
        total_expenses = float(tax) + float(insurance) + float(utilities) + float(hoa) + float(maintenance) + float(vacancy) + float(repairs) + float(capex) + float(prop_man) + float(mortgage)
        self.monthly_expenses = "{:.2f}".format(total_expenses)
        print(f"\nTotal Monthly Expenses: ${self.monthly_expenses}.\n")

    
    def cashflow(self):
        total_cashflow = float(self.monthly_income) - float(self.monthly_expenses)
        self.monthly_cashflow = "{:.2f}".format(total_cashflow)
        print(f"\nMonthly Cash Flow: ${self.monthly_cashflow}.\n")

    def annual_ROI(self):
        down_payment = input("What was the total down payment? ")
        closing_cost = input("How much were closing costs? ")
        rehab = input("How much was spent on initial renovations? ")
        misc = input("Any other one-time expenses for the property? ")
        total_investment = float(down_payment) + float(closing_cost) + float(rehab) + float(misc)
        yearly_cashflow = float(self.monthly_cashflow) * 12
        roi = yearly_cashflow / total_investment * 100
        self.roi_percent = "{:.2f}".format(roi)

        print(f"\nMonthly Income: ${self.monthly_income}")
        print(f"Monthly Expenses: ${self.monthly_expenses}")
        print(f"Monthly Cash Flow: ${self.monthly_cashflow}")
        print(f"\nANNUAL ROI: {self.roi_percent}%")


rentalProperty = ROI_Calc()

def run():
    while True:
        rentalProperty.income()
        rentalProperty.expenses()
        rentalProperty.cashflow()
        rentalProperty.annual_ROI()
        break
run()
```
